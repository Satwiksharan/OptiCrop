```python
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 20)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (12, 8)

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
from ipywidgets import interact

from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Load the agricultural dataset
data = pd.read_csv('data/Crop_recommendation.csv')

# Display the initial records to verify structure
data.head()
```

```python
# Plotting distributions of environmental and soil conditions
plt.subplot(2, 4, 1)
sns.histplot(data['N'], color='orange', kde=True)
plt.xlabel('Ratio of Nitrogen', fontsize=12)
plt.grid()

plt.subplot(2, 4, 2)
sns.histplot(data['P'], color='blue', kde=True)
plt.xlabel('Ratio of Phosphorous', fontsize=12)
plt.grid()

plt.subplot(2, 4, 3)
sns.histplot(data['K'], color='darkblue', kde=True)
plt.xlabel('Ratio of Potassium', fontsize=12)
plt.grid()

plt.subplot(2, 4, 4)
sns.histplot(data['temperature'], color='green', kde=True)
plt.xlabel('Temperature', fontsize=12)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis: Humidity vs Crop Type
plt.figure(figsize=(12, 8))
sns.scatterplot(x=data['humidity'], y=data['label'], color='teal')
plt.title('Humidity Distribution across Different Crops', fontsize=16)
plt.xlabel('Humidity (%)', fontsize=12)
plt.ylabel('Crop Label', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Multivariate Analysis: Target Label Count Distribution
plt.figure(figsize=(15, 7))
sns.countplot(x='label', data=data, palette='Set2')
plt.title('Distribution of Agricultural Crop Labels', fontsize=16)
plt.xlabel('Crop Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Checking dataset dimensions and missing integrity parameters
print("Dataset Shape:", data.shape)
print("\nDataset Info:")
data.info()

# Explicit summation of null records per column
print("\nMissing Values Breakdown:")
print(data.isnull().sum())
```

```python
# Visualizing outliers using Boxplots
plt.figure(figsize=(8, 4))
sns.boxplot(x=data['K'], color='purple')
plt.title('Identifying Outliers in Potassium (K) Feature')
plt.show()

# Calculating IQR Bounds for Potassium (K)
Q1 = data['K'].quantile(0.25)
Q3 = data['K'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print(f"Potassium Outlier Bounds: Upper={upper_bound}, Lower={lower_bound}")

# Applying Log Transformation to stabilize variance and handle skewness
data['K'] = np.log1p(data['K'])
print("Log transformation applied to the Potassium ('K') feature successfully.")
```

```python
# Extracting Summer Crops
print("Summer crops")
print(data[(data['temperature'] > 30) & (data['humidity'] > 50)]['label'].unique())
print("---------------------------------------------------------")

# Extracting Winter Crops
print("Winter crops")
print(data[(data['temperature'] < 20) & (data['humidity'] > 30)]['label'].unique())
print("---------------------------------------------------------")

# Extracting Rainy Crops
print("Rainy crops")
print(data[(data['rainfall'] > 200) & (data['humidity'] > 50)]['label'].unique())
print("---------------------------------------------------------")
```

```python
# Separating features and target variable
y = data['label']
x = data.drop(['label'], axis=1)

print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
print("---------------------------------------------------------")

# Splitting the dataset into Training and Testing sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

print("The shape of x train:", x_train.shape)
print("The shape of x test:", x_test.shape)
print("The shape of y train:", y_train.shape)
print("The shape of y test:", y_test.shape)
```

```python
# Finding the optimal number of clusters using the Elbow Method
plt.rcParams['figure.figsize'] = (10, 4)
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)

# Plotting the Elbow Graph
plt.plot(range(1, 11), wcss, marker='o', color='green')
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('No. of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# Initializing and training the Logistic Regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000, random_state=0)
model.fit(x_train, y_train)

# Generating predictions on the testing dataset
y_pred = model.predict(x_test)
print("Logistic Regression model trained and predictions generated successfully.")
```
